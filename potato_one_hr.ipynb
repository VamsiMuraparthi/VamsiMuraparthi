```python
import pandas as pd
input_file = "/content/211McCain_one_hourly_aggregate_data.csv"
df = pd.read_csv(input_file,skiprows=[1])
```

```python
df2=pd.DataFrame()
```

```python
df.info()
df2["Date"]=df["Date"]
df2["Time_Slot"]=df["Time_Slot"]
df2["Ambient relative humidity %RH"]=df["Ambient relative humidity %RH"]
df2["Ambient temperature Deg Centigrade"]=df["Ambient temperature Deg Centigrade"]
```

```python
print(df2.head())
df2.reset_index(inplace=False)
print(df2.head())
```

```python
df2.to_csv("temp.csv",index=False)
```

```python
from datetime import datetime,timedelta
df2["Date"] = pd.to_datetime(df2["Date"],format='%d-%b-%y') #'%d-%m-%Y') #'%Y-%m-%d')  #'%d-%m-%Y') #'%Y-%m-%d') #format='%Y-%m-%d')
```

```python
start_date=df2.loc[0,"Date"]
start_time=df2.loc[0,"Time_Slot"]
print(start_date,"time=",start_time)
df2.loc[0,"hours"]=0
for i in range(1,len(df2)) :
    date2=df2.loc[i,"Date"]
    diff=date2-start_date;
    d=diff.days
    d=d-1
    if(d<0):
        d=0
    hours=int(d)*24;
    print(int(hours))
    start_date_hours=24-start_time
    end_date_hours=df2.loc[i,"Time_Slot"]
    if(date2>start_date):
        df2.loc[i,"hours"]=hours+start_date_hours+end_date_hours
    else:
        df2.loc[i,"hours"]=end_date_hours-start_time
```
